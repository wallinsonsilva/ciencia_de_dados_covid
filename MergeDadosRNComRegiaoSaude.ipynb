```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
```

```
acumulado_rn = pd.read_csv('/content/drive/My Drive/Projeto Ciencia de dados - COVID 19/Dados Gerados/ET_CASOS_ACUMULADO_RN.csv')
acumulado_rn.head()
```

```
acumulado_rn.describe()
```

```
regioes_saude = pd.read_csv('/content/drive/My Drive/Projeto Ciencia de dados - COVID 19/CIR RN/MUNICIPIOS POR REGIAO SAUDE RN.csv')
regioes_saude.CIR = regioes_saude.CIR.astype(str)
regioes_saude.CodRegiao = regioes_saude.CodRegiao.astype(str)
regioes_saude.head()
```

```
regioes_saude.isnull().sum()
```

```
casos_mun_reg = pd.merge(acumulado_rn, regioes_saude, left_on='codmun',right_on='CodMunIBGE')
casos_mun_reg.head()
```

```
casos_mun_reg.to_csv('/content/drive/My Drive/Projeto Ciencia de dados - COVID 19/Dados Gerados/ET_casos_municipios_regiao.csv',index=False)
```

```
dados_ibge_rn = pd.read_excel('/content/drive/My Drive/Projeto Ciencia de dados - COVID 19/CIR RN/Dado IGBE 2017.xlsx')
dados_ibge_rn.head()
```

```
dados_ibge_rn.isnull().sum() 
```

```
#Tratando valores nulos sem remover
dados_ibge_rn['Receitas realizadas - R$ (×1000)'] = dados_ibge_rn['Receitas realizadas - R$ (×1000)'].fillna(dados_ibge_rn['Receitas realizadas - R$ (×1000)'].mean())
dados_ibge_rn['Despesas empenhadas - R$ (×1000) [2017]'] = dados_ibge_rn['Despesas empenhadas - R$ (×1000) [2017]'].fillna(dados_ibge_rn['Despesas empenhadas - R$ (×1000) [2017]'].mean())
dados_ibge_rn.isnull().sum()
```

```
dados_ibge_rn = dados_ibge_rn.drop(['Município','Gentílico','Prefeito'],axis=1)
dados_ibge_rn.head()
```

```
dados_ibge_rn.info()
```

```
dados_casos_mun_reg_ibge = pd.merge(casos_mun_reg,dados_ibge_rn,left_on='Código Município Completo',right_on='Código')
dados_casos_mun_reg_ibge.head()
```

```
dados_casos_mun_reg_ibge.to_csv('/content/drive/My Drive/Projeto Ciencia de dados - COVID 19/Dados Gerados/ET_dados_casos_mun_reg_igbe.csv',index=False)
```
